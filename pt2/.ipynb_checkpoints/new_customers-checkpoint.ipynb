```python
#invite people for the Kaggle party
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('new_customers_update.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
dob = df['DOB']
```

```python
years = []

for i in dob:
    years.append(i[:4])
```

```python
df['yob'] = years
```

```python
df = df.drop(columns=['DOB'])
```

```python
df['yob'] = df['yob'].astype('int')
```

```python
decade = []

for index, row in df.iterrows():
    if (row["yob"] < 1950):
        decade.append('40')
    elif ((row["yob"] < 1960) and (row['yob'] > 1949)):
        decade.append('50')
    elif ((row["yob"] < 1970) and (row['yob'] > 1959)):
        decade.append('60')
    elif ((row["yob"] < 1980) and (row['yob'] > 1969)):
        decade.append('70')
    elif ((row["yob"] < 1990) and (row['yob'] > 1979)):
        decade.append('80')
    elif ((row["yob"] < 2000) and (row['yob'] > 1989)):
        decade.append('90')
    elif (row["yob"] > 1999):
        decade.append('00')
    else: 
        print(row['yob'])
```

```python
df['decade'] = decade
```

```python
df['decade'] = df['decade'].astype('int')
```

```python
df.corr()
```

```python
corr = df.corr()
sns.heatmap(corr, cmap="Blues", annot=True)
```
